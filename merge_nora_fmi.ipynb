```python
# Data for each NORA station
data = {}
time_station1 = pd.to_datetime(dfs_fmi[0].index.tolist())
time_station2 = pd.to_datetime(dfs_fmi[3].index.tolist())
time_station3 = pd.to_datetime(dfs_fmi[6].index.tolist())
frequency = dfs_fmi[0].columns.tolist()
paramind = 0


for station, time in zip(stations, [time_station1, time_station2, time_station3]):
    data[station] = xr.Dataset(
        {
            "insitu_mdir": xr.DataArray(
                dfs_fmi[paramind],
                dims=("time", "frequency"),
                coords={"time": time, "frequency": frequency},
            ),
            "insitu_spec": xr.DataArray(
                dfs_fmi[paramind+1],
                dims=("time", "frequency"),
                coords={"time": time, "frequency": frequency},
            ),
            "insitu_sprd": xr.DataArray(
                dfs_fmi[paramind+2],
                dims=("time", "frequency"),
                coords={"time": time, "frequency": frequency},
            ),
        }
        
    )
    paramind = paramind + 3
```
