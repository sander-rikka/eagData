```python
import glob
import xarray as xr
import pandas as pd
import numpy as np
```

```python
fmi_path = r'data\FMI'
nora_path = r'data\NORA'
fmi_files = glob.glob(rf'{fmi_path}\*.csv')
nora_files = glob.glob(rf'{nora_path}\*.csv')

stations = ['bot','gof','nbp']
params = ['mdir', 'spec', 'spr']

dfs_nora = [pd.read_csv(x, index_col=None) for x in nora_files]
dfs_fmi =  [pd.read_csv(x, index_col=None) for x in fmi_files]
```

```python
# Data for each NORA station
data = {}
time_station1 = pd.to_datetime(dfs_nora[0].index.tolist())
time_station2 = pd.to_datetime(dfs_nora[3].index.tolist())
time_station3 = pd.to_datetime(dfs_nora[6].index.tolist())
frequency = dfs_nora[0].columns.tolist()
paramind = 0


for station, time in zip(stations, [time_station1, time_station2, time_station3]):
    data[station] = xr.Dataset(
        {
            "nora_mdir": xr.DataArray(
                dfs_nora[paramind],
                dims=("time", "frequency"),
                coords={"time": time, "frequency": frequency},
            ),
            "nora_spec": xr.DataArray(
                dfs_nora[paramind+1],
                dims=("time", "frequency"),
                coords={"time": time, "frequency": frequency},
            ),
            "nora_sprd": xr.DataArray(
                dfs_nora[paramind+2],
                dims=("time", "frequency"),
                coords={"time": time, "frequency": frequency},
            ),
        }
        
    )
    paramind = paramind + 3
```

```python
# Step 1: Create a union of all time indexes
union_time = pd.to_datetime(
    sorted(
        set(time_station1).union(set(time_station2)).union(set(time_station3))
    )
)
```

```python
# Step 2: Reindex each station's dataset to the union time index
aligned_data = {}
for station, ds in data.items():
    aligned_data[station] = ds.reindex(time=union_time)
```

```python
# Step 3: Add station dimension and combine all stations
datasets = []
for station, ds in aligned_data.items():
    ds = ds.expand_dims(station=[station])  # Add station dimension
    datasets.append(ds)
```

```python
# Merge all datasets along the station dimension
fmi_data = xr.concat(datasets, dim="station")
```

```python
nora_data = xr.concat(datasets, dim="station")
```
